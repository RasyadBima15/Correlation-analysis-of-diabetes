```python
# import library
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# mendapatkan  correlations dari setiap fitur dalam dataset
data = pd.read_csv("diabetes.csv")
corrmat = data.corr()
top_corr_features = corrmat.index
```

```python
# plot heatmap
plt.figure(figsize=(10,10))
g=sns.heatmap(data[top_corr_features].corr(),annot=True,cmap="RdYlGn")
```

Dalam kasus di atas, saya ingin menganalisis tingkat diabetes pada dataset yang diberikan dengan melihat variabel Outcome yang menandakan apabila nilainya 1, maka menunjukkan bahwa individu tersebut memiliki diabetes dan apabila 0 maka individu tersebut tidak memiliki diabetes. Untuk menganalisis variabel-variabel apa saja yang mempengaruhi tingkat diabetes, saya menggunakan library matplotlib untuk mengetahui korelasi antara variabel bebas dengan variabel terikat.

ket :

      variabel bebas -> selain variabel Outcome
      variabel terikat -> variabel Outcome

Berikut ini hasil analisis korelasi antara variabel Outcome dengan variabel lainnya.

1. Outcome dengan Age (Usia): Korelasi sebesar 0.24 menunjukkan adanya hubungan positif yang moderat antara usia dan keberadaan diabetes. Ini mengindikasikan bahwa seiring dengan bertambahnya usia, risiko seseorang memiliki diabetes meningkat, meskipun korelasinya tidak terlalu kuat.

2. Outcome dengan DiabetesPedigreeFunction: Dengan korelasi sebesar 0.17, ada hubungan positif yang lemah antara fungsi keturunan diabetes dengan outcome. Ini menandakan bahwa individu dengan riwayat keluarga yang memiliki diabetes memiliki kemungkinan sedikit lebih besar untuk mengembangkan diabetes.

3. Outcome dengan BMI: Korelasi sebesar 0.29 antara BMI dan outcome mengindikasikan hubungan positif yang moderat. Individu dengan BMI yang lebih tinggi cenderung memiliki risiko yang lebih tinggi untuk diabetes.

4. Outcome dengan Insulin: Korelasi 0.13 menunjukkan hubungan yang lemah antara kadar insulin dalam darah dan keberadaan diabetes. Meskipun demikian, ini tetap memberikan bukti bahwa kadar insulin dapat mempengaruhi risiko diabetes Peningkatan kadar insulin menunjukkan resistensi insulin, di mana tubuh tidak merespons insulin dengan efektif. Resistensi insulin bisa meningkatkan risiko diabetes.

5. Outcome dengan SkinThickness: Dengan korelasi sebesar 0.075, hubungan antara ketebalan kulit dan outcome sangat lemah, menunjukkan bahwa faktor ini mungkin kurang relevan dalam prediksi diabetes dibandingkan variabel lain.

6. Outcome dengan BloodPressure: Korelasi sebesar 0.065 juga menunjukkan hubungan yang sangat lemah antara tekanan darah dan keberadaan diabetes.

7. Outcome dengan Glucose: Dengan korelasi sebesar 0.47, ini adalah salah satu korelasi terkuat dalam analisis ini. Ini menunjukkan bahwa kadar glukosa dalam darah memiliki hubungan yang kuat dengan keberadaan diabetes. Semakin tinggi kadar glukosa dalam darah, risiko diabetes juga meningkat.

8. Outcome dengan Pregnancies: Korelasi sebesar 0.22 menunjukkan hubungan positif yang moderat antara jumlah kehamilan dengan outcome. Hal ini menunjukkan bahwa wanita yang telah mengalami lebih banyak kehamilan mungkin memiliki risiko yang sedikit lebih besar terhadap diabetes.

**Kesimpulan**

Berdasarkan hasil analisis korelasi di atas, menunjukkan variabel yang memiliki korelasi terkuat dengan variabel outcome adalah variabel glukosa. kadar glukosa dalam darah memiliki hubungan yang kuat dengan keberadaan diabetes. **Jadi dapat disimpulkan, semakin tinggi kadar glukosa dalam darah, risiko diabetes juga meningkat. Begitupun sebaliknya, semakin rendah kadar glukosa dalam darah, risiko diabetes menurun.**
